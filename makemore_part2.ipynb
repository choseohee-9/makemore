```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
words = open('names.txt', 'r').read().splitlines()
words[:8]
```

```python
len(words)
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}

print(itos)
```

```python
block_size = 3

X, Y = [], []

for w in words:

    context = [0] * block_size

    for ch in w + '.':
        ix = stoi[ch]

        X.append(context)
        Y.append(ix)

        context = context[1:] + [ix]

X = torch.tensor(X)
Y = torch.tensor(Y)

X.shape, Y.shape
```

```python
C = torch.randn((27,2))

emb = C[X]

emb.shape
```

```python
W1 = torch.randn((6,100))
b1 = torch.randn(100)

h = torch.tanh(
    emb.view(-1,6) @ W1 + b1
)

h.shape
```

```python
W2 = torch.randn((100,27))
b2 = torch.randn(27)

logits = h @ W2 + b2

logits.shape
```

```python
counts = logits.exp()

prob = counts / counts.sum(1, keepdims=True)

loss = -prob[torch.arange(32), Y[:32]].log().mean()

loss
```
